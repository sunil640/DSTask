# Preprocessing & Feature Scaling

```python
#scaling of dataset
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
dfs=ss.fit_transform(df[['X55','X56','X57']])
```

```python
dfs
```

```python
dfs=pd.DataFrame(dfs)
```

```python
dfs.head()
```

```python
#concating scale dataset and origional dataset
df=pd.concat([dfs,df],axis=True)
```

```python
df.head()
```

```python
#droping unscaled columns
df.drop(['X55','X56','X57'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#renaming the scaled dataset column
df.columns=['X55','X56','X57','Y']
```

```python
df.head()
```

# Spliting the dataset for training and Validation

```python
#selecting features and targate
x=df.iloc[:,:3]
y=df.iloc[:,-1]
```

```python
#spliting the data set into training and validation data
from sklearn.model_selection import train_test_split
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=2,test_size=0.2)
```

# 1st Model Creation

```python
#Model creation
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

# Eavaluation of model1

```python

from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

# 2nd Model Creation

```python
#checking and eavaluation of with other model
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypredrf=rf.predict(xtest)
```

# Eavaluation of model2

```python
accuracy_score(ytest,ypredrf)
```

```python
print(classification_report(ytest,ypredrf))
```

# Hyperparameter tuning and eavaluation of model2

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param={'n_estimators':[100],
       'criterion':['gini','entropy'],
       'min_samples_split':[2],
       'min_samples_leaf':[2],
       'max_features':['sqrt'],
       'ccp_alpha':[0.1],
       'verbose':[0],
       'max_depth':[4,6]
       }
grid=GridSearchCV(rf,param,scoring='accuracy',cv=5)
grid.fit(xtrain,ytrain)
```

```python
print(grid.best_params_)
```

```python
#retraining the model with with best parameter
rfc=RandomForestClassifier(ccp_alpha= 0.1, criterion= 'entropy', max_depth= 6, max_features= 'sqrt', min_samples_leaf= 2, min_samples_split= 2, n_estimators =100, verbose= 0)
rfc.fit(xtrain,ytrain)
ypredrf=rfc.predict(xtest)
```

```python
#checking perfomance of model
accuracy_score(ytest,ypredrf)
```

```python
print(classification_report(ytest,ypredrf))
```
